```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path = '/content/train.csv'
titanic = pd.read_csv(path)
titanic.head()  #first 5 rows
titanic.info()  #datatypes / info
```

```python
#Survival By Gender
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='Sex', hue='Survived') #countplot
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
#count by passenger class
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='Pclass', hue='Survived', palette='Set3') #countplot
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# By Age
plt.figure(figsize=(8, 5))
sns.histplot(titanic['Age'], kde=True, color='grey') #histogram
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(titanic['Fare'], kde=True, bins=30, color='green')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
